```python
ii = np.asarray([0, 0, 0, 1, 1, 1, 2, 2])
jj = np.asarray([4, 5, 1, 6, 7, 2, 3, 8])
natoms = 9

#do the above with the brute sel minimizers
from numpy import asarray as array
sel_brute = {0: array([712, 732]),
 1: array([718, 732]),
 2: array([718, 732]),
 3: array([712, 732]),
 4: array([712, 732]),
 5: array([718, 732]),
 6: array([718, 727]),
 7: array([712, 732]),
 8: array([715, 732]),
 9: array([712, 732]),
 10: array([718, 732]),
 11: array([712, 732]),
 12: array([712, 738]),
 13: array([712, 739]),
 14: array([712, 732]),
 15: array([712, 732]),
 16: array([712, 732]),
 17: array([712, 732]),
 18: array([718, 738]),
 19: array([718, 732]),
 20: array([718, 732]),
 21: array([712, 732]),
 22: array([712, 732]),
 23: array([712, 738]),
 24: array([718, 732])}

with open('/Users/samsonkoelle/manifoldflasso_jmlr/untracked_data/embeddings/malonaldehyde/malonaldehyde_0223_fullreplicates_small.pkl' ,
         'rb') as loader:
     replicates = pickle.load(loader)
        
nreps = 25
sel_l = np.zeros(nreps, dtype = int)
sel = {}
for r in range(nreps):
    cs = rearrange(np.asarray(list(replicates[r].results[1].values())), 'l n p m -> l m n p')
    xaxis = np.asarray(np.asarray(list(replicates[r].results[1].keys())))
    xaxis_reorder = xaxis[xaxis.argsort()]
    cs_reorder = cs[xaxis.argsort()]
    
    replicates[r].cs = cs
    replicates[r].cs_reorder = cs_reorder
    replicates[r].xaxis_reorder = xaxis_reorder
    replicates[r].xaxis = xaxis
    
    sel_l[r] = np.where(xaxis_reorder ==  replicates[r].results[0])[0][0]
    sel[r] = np.where(np.linalg.norm(replicates[r].cs_reorder, axis = tuple([1,2,]))[sel_l[r]] !=0.)[0]
    
atoms4,p = get_atoms_4(natoms, ii, jj)
atoms4_all,p = get_all_atoms_4(natoms)
probe = atoms4[0]
superset = atoms4_all
superset = np.asmatrix(superset)

def get_index_matching(probe, superset):
    
    probe_permuted = np.asarray([probe[[0, 1, 2, 3]],
                                 probe[[3,2,1,0]],
                                 # combos[c][[0,2,1,3]],
                                 probe[[0,2,1,3]],
                                 probe[[3,1,2,0]]])
    
    output = np.asarray([])
    for p in range(4):
        #print(p)
        output = np.append(output,np.where((superset==tuple(probe_permuted[p])).all(1))[0])

    return(int(output))

matchingindex= np.zeros(12)
for p in range(12):
    matchingindex[p] = get_index_matching(atoms4[p], superset)
matchingindex = np.asarray(matchingindex, dtype = int)
```

```python
#0 lambda true coeffs
tol = 1e-16
learning_rate = 1.
max_iter = 500
reg_l2 = 0

l1norm = np.zeros((nreps,12,12))
penalties =np.zeros((nreps,12,12))
l2losses = np.zeros((nreps,12,12))
losses = np.zeros((nreps,12,12))
for r in range(nreps):
    print(r)
    rep = replicates[r]
    for i in range(6):
        for j in range(6,12):
            #true_sup = #np.asarray([0,8])
            dg_M = rep.dg_M[:,:,[matchingindex[i], matchingindex[j]]]
            df_M = rep.df_M
            reg_l1s = np.asarray([0.])
            GGL0 = GradientGroupLasso(dg_M, df_M, reg_l1s, reg_l2, max_iter,learning_rate, tol)
#             GGL0.fit()
#             beta_npm = GGL0.fit_[-1]['beta']
            beta_npm = np.asarray([np.linalg.lstsq(dg_M[i], df_M[i])[0] for i in range(100)])
            l1norm[r,i,j] = GGL0._L1penalty(beta_npm)
            penalties[r,i,j] = GGL0._L1penalty(beta_npm) * reg_l1s[0]
            l2losses[r,i,j] = GGL0._L2loss(beta_npm)
            losses[r,i,j] = GGL0._L1penalty(beta_npm) * reg_l1s[0] + GGL0._L2loss(beta_npm)
```
